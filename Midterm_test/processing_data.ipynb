```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df =  pd.read_csv('./data-kiem-tra-2.csv')
df.head()
```

```python
def sort_data(df):
    df = df.sort_values(['date'])
    return df

df = sort_data(df.copy())
df.head()
```

```python
df.info()
```

```python
def mean_day_data(df):
    df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
    df_sum = df.groupby('date').mean().reset_index()
    return df_sum
df_sum = mean_day_data(df.copy())
df_sum.head(10)
```

```python
# Trc quan ha tng ct
plt.figure(figsize=(15, 10))
plt.subplot(3, 2, 1)
plt.plot(df_sum['date'], df_sum['truong_1'])
plt.title('truong_1')
plt.subplot(3, 2, 2)
plt.plot(df_sum['date'], df_sum['truong_2'])
plt.title('truong_2')
plt.subplot(3, 2, 3)
plt.plot(df_sum['date'], df_sum['truong_3'])
plt.title('truong_3')
plt.subplot(3, 2, 4)
plt.plot(df_sum['date'], df_sum['truong_4'])
plt.title('truong_4')
plt.subplot(3, 2, 5)
plt.plot(df_sum['date'], df_sum['truong_5'])
plt.title('truong_5')
plt.tight_layout()
plt.show()
```
